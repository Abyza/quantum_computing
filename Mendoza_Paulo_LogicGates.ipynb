# Assignment

Implement classic logic gates using quantum gates
* XOR
* OR
* AND
* XNOR
* NOR
* NAND

```python
print('XNor Gate')

circuit = QuantumCircuit(2)

circuit.cx(0, 1)
circuit.x(1)

circuit.draw('mpl')
```

```python
qi.Statevector.from_instruction(circuit).draw('latex')
```

```python
print('Xor Gate')

circuit = QuantumCircuit(2)

circuit.cx(0, 1)

circuit.draw('mpl')
```

```python
qi.Statevector.from_instruction(circuit).draw('latex')
```

```python
print('Or Gate')

circuit = QuantumCircuit(3)

circuit.x([0, 1])
circuit.ccx(0, 1, 2)
circuit.x([0, 1, 2])

circuit.draw('mpl')
```

```python
print('And Gate')

circuit = QuantumCircuit(3)

circuit.x([0, 1])
circuit.ccx(0, 1, 2)
circuit.x([0, 1, 2])

circuit.draw('mpl')
```
