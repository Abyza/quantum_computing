```python
from qiskit import *
import qiskit.quantum_info as qi
from qiskit_aer import Aer
from qiskit.visualization import plot_histogram
```

Implement classic logic gates using quantum gates

* OR
* NOR
* AND
* NAND
* XOR
* XNOR

```python
print('Or Gate')

or_circuit = QuantumCircuit(3, 3)

or_circuit.barrier()
or_circuit.cx(0, 2)
or_circuit.cx(1, 2)
or_circuit.ccx(0, 1, 2)
or_circuit.barrier()

or_circuit.draw('mpl')
```

```python
print('Or Gate State vector')
qi.Statevector.from_instruction(or_circuit).draw('latex')
```

```python
print('Or Gate Testing')

or_circuit = QuantumCircuit(3, 3)

or_circuit.x([0,1]) # change this to test different inputs
or_circuit.barrier()
or_circuit.cx(0, 2)
or_circuit.cx(1, 2)
or_circuit.ccx(0, 1, 2)
or_circuit.barrier()
or_circuit.measure(2, 2) # the first 2 is the qubit and the send 2 is the classical bit 
#this starts at 0 for both of them so it measures the 3 qbit and the most significant classical bit is the output of the logic gate

or_circuit.draw('mpl')
```

```python
print('Or Gate Testing Result')

backend = Aer.get_backend('qasm_simulator')
job = backend.run(or_circuit, shots = 100000)
result = job.result()
plot_histogram(result.get_counts(or_circuit))
```
