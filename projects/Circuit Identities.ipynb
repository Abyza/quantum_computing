## Part 1

### DELAY Gate

```python
print('Delay Gate')

not_circuit = QuantumCircuit(1,1)

not_circuit.barrier()
not_circuit.h(0)
not_circuit.h(0)
not_circuit.barrier()

not_circuit.draw('mpl')
```

```python
print('Delay Gate State vector')
qi.Statevector.from_instruction(not_circuit).draw('latex')
```

```python
print('Delay Gate Testing')

not_circuit = QuantumCircuit(1,1)

not_circuit.x([0]) # change this to test different inputs
not_circuit.barrier()
not_circuit.h(0)
not_circuit.h(0)
not_circuit.barrier()
not_circuit.measure(0,0) # the first 0 is the qubit and the send 0 is the classical bit 
#this starts at 0 for both of them so it measures the 1 qbit and the most significant classical bit is the output of the logic gate

not_circuit.draw('mpl')
```

```python
print('Delay Gate Testing Result')

backend = Aer.get_backend('qasm_simulator')
job = backend.run(not_circuit, shots = 100000)
result = job.result()
plot_histogram(result.get_counts(not_circuit))
```

```python
print('Delay Gate Testing All')

two_inputs = [[],[0]] # the 1st entry [] means that no not gate is added but [0] means not gate is added at q[0]

for i in two_inputs:
    not_circuit = QuantumCircuit(1, 1)
    if not i:
        not_circuit.barrier()
        not_circuit.h(0)
        not_circuit.h(0)
        not_circuit.barrier()
        not_circuit.measure(0, 0) 

        backend = Aer.get_backend('qasm_simulator')
        job = backend.run(not_circuit, shots = 100000)
        result = job.result()
        print([],'\t',result.get_counts())
        
    else: 
        not_circuit.x(i) 
        not_circuit.barrier()
        not_circuit.h(0)
        not_circuit.h(0)
        not_circuit.barrier()
        not_circuit.measure(0, 0) 

        backend = Aer.get_backend('qasm_simulator')
        job = backend.run(not_circuit, shots = 100000)
        result = job.result()
        print(i,'\t',result.get_counts())
```

###  Full-Adder circuit

```python
print('Full-Adder Gate')

# Input
# q[0] - A
# q[1] - B
# q[2] - CarryIn
# q[3] - not used

# Output
# q[0] - A
# q[1] - B
# q[2] - SumOut
# q[3] - CarryOut

full_adder_circuit = QuantumCircuit(4, 4)

full_adder_circuit.barrier()
full_adder_circuit.ccx(0, 1, 3)
full_adder_circuit.cx(0, 1)
full_adder_circuit.ccx(1, 2, 3)
full_adder_circuit.cx(1, 2)
full_adder_circuit.cx(0, 1)
full_adder_circuit.barrier()

full_adder_circuit.draw('mpl')
```

```python
qi.Statevector.from_instruction(full_adder_circuit).draw('latex')
```

```python
print('Full-Adder Gate')

full_adder_circuit = QuantumCircuit(4, 4)

full_adder_circuit.x([0,1,2])
full_adder_circuit.barrier()
full_adder_circuit.ccx(0, 1, 3)
full_adder_circuit.cx(0, 1)
full_adder_circuit.ccx(1, 2, 3)
full_adder_circuit.cx(1, 2)
full_adder_circuit.cx(0, 1)
full_adder_circuit.barrier()
full_adder_circuit.measure(0, 0)
full_adder_circuit.measure(1, 1)
full_adder_circuit.measure(2, 2)
full_adder_circuit.measure(3, 3)

full_adder_circuit.draw('mpl')
```

```python
backend = Aer.get_backend('qasm_simulator')
job = backend.run(full_adder_circuit, shots = 100000)
result = job.result()
plot_histogram(result.get_counts())
```

```python
print('Full-Adder Gate Testing All')

full_adder_inputs = [[],[2],[1],[1,2],[0],[0,2],[0,1],[0,1,2]]

for i in full_adder_inputs:
    full_adder_circuit = QuantumCircuit(4, 4)
    if not i:
        full_adder_circuit.barrier()
        full_adder_circuit.ccx(0, 1, 3)
        full_adder_circuit.cx(0, 1)
        full_adder_circuit.ccx(1, 2, 3)
        full_adder_circuit.cx(1, 2)
        full_adder_circuit.cx(0, 1)
        full_adder_circuit.barrier()
        full_adder_circuit.measure(0, 0)
        full_adder_circuit.measure(1, 1)
        full_adder_circuit.measure(2, 2)
        full_adder_circuit.measure(3, 3)

        backend = Aer.get_backend('qasm_simulator')
        job = backend.run(full_adder_circuit, shots = 100000)
        result = job.result()
        print([],'\t',result.get_counts())
        
    else: 
        full_adder_circuit.x(i) 
        full_adder_circuit.barrier()
        full_adder_circuit.ccx(0, 1, 3)
        full_adder_circuit.cx(0, 1)
        full_adder_circuit.ccx(1, 2, 3)
        full_adder_circuit.cx(1, 2)
        full_adder_circuit.cx(0, 1)
        full_adder_circuit.barrier()
        full_adder_circuit.measure(0, 0)
        full_adder_circuit.measure(1, 1)
        full_adder_circuit.measure(2, 2)
        full_adder_circuit.measure(3, 3)

        backend = Aer.get_backend('qasm_simulator')
        job = backend.run(full_adder_circuit, shots = 100000)
        result = job.result()
        print(i,'\t',result.get_counts())
```
