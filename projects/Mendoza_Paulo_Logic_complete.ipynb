# Assignment

Implement classic logic gates using quantum gates
* XOR
* OR
* AND
* XNOR
* NOR
* NAND

```python
print('XNor Gate')

xnor_circuit = QuantumCircuit(2, 2)

xnor_circuit.cx(0, 1)
xnor_circuit.x(1)
xnor_circuit.measure(1, 1)

xnor_circuit.draw('mpl')
```

```python
qi.Statevector.from_instruction(circuit).draw('latex')
```

```python
print('Xor Gate')

xor_circuit = QuantumCircuit(2, 2)

xor_circuit.cx(0, 1)
xor_circuit.measure(1, 1)

xor_circuit.draw('mpl')
```

```python
print('Or Gate')

or_circuit = QuantumCircuit(3)

or_circuit.x([0, 1])
or_circuit.ccx(0, 1, 2)
or_circuit.x([0, 1, 2])

or_circuit.draw('mpl')
```

```python
print('Or Gate')

or_circuit = QuantumCircuit(3, 3)

or_circuit.barrier(0)
or_circuit.x(1)
or_circuit.cx(0, 2)
or_circuit.barrier(1)
or_circuit.cx(1, 2)
or_circuit.ccx(0, 1, 2)
or_circuit.measure(2, 2)

or_circuit.draw('mpl')
```

```python
print('Nor Gate')

nor_circuit = QuantumCircuit(3, 3)

nor_circuit.barrier(0)
nor_circuit.x(1)
nor_circuit.cx(0, 2)
nor_circuit.barrier(1)
nor_circuit.cx(1, 2)
nor_circuit.ccx(0, 1, 2)
nor_circuit.x(2)
nor_circuit.measure(2, 2)

nor_circuit.draw('mpl')
```

```python
print('And Gate')

and_circuit = QuantumCircuit(3, 3)

and_circuit.x([0, 1])
and_circuit.barrier([0, 1])
and_circuit.ccx(0, 1, 2)
and_circuit.barrier()
and_circuit.measure(2, 2)

and_circuit.draw('mpl')
```

```python
print('Nand Gate')

nand_circuit = QuantumCircuit(3, 3)

nand_circuit.ccx(0, 1, 2)
nand_circuit.barrier()
nand_circuit.x(2)
nand_circuit.barrier(2)
nand_circuit.measure(2, 2)


nand_circuit.draw('mpl')
```
